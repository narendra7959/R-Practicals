```r
# Load necessary libraries
install.packages("ggplot2")  # Install if not already installed
library(ggplot2)
```

```r
# Load the iris dataset
data("iris")
```

```r
# Step 1: Preprocess the Data
# Remove categorical column and keep only numerical features
iris_numeric <- iris[, 1:4]
```

```r
# Check for missing values
sum(is.na(iris_numeric))  # Should be 0 for a clean dataset
```

```r
# Step 2: Standardization (Mean = 0, Variance = 1)
iris_scaled <- scale(iris_numeric)  # Standardizing the data
```

```r
# Step 3: Compute PCA
pca_result <- prcomp(iris_scaled, center = TRUE, scale. = TRUE)
```

```r

# Step 4: Extract PCA details
summary(pca_result)  # Variance explained by each principal component
pca_result$rotation  # Principal components (loadings)
pca_result$x  # Transformed dataset (scores)
pca_result$sdev^2  # Eigenvalues (variance of each principal component)
```

```r
# Step 5: Scree Plot (Variance Explained by Each Principal Component)
variance_explained <- pca_result$sdev^2 / sum(pca_result$sdev^2) * 100
scree_df <- data.frame(PC = paste0("PC", 1:length(variance_explained)), Variance = variance_explained)

ggplot(scree_df, aes(x = PC, y = Variance)) +
  geom_bar(stat = "identity", fill = "steelblue", color = "black") +
  geom_text(aes(label = round(Variance, 1)), vjust = -0.5, size = 5) +
  labs(title = "Scree Plot (Variance Explained)", x = "Principal Components", y = "Percentage of Variance") +
  theme_minimal()
```

```r
# Step 6: Correlation Between Original Features
cor_matrix <- cor(iris_numeric)
cor_df <- as.data.frame(as.table(cor_matrix))

ggplot(cor_df, aes(Var1, Var2, fill = Freq)) +
  geom_tile(color = "white") +
  scale_fill_gradient2(low = "blue", high = "red", mid = "white", midpoint = 0, limit = c(-1, 1), space = "Lab") +
  geom_text(aes(label = round(Freq, 2)), color = "black", size = 4) +
  labs(title = "Feature Correlation Matrix", x = "", y = "") +
  theme_minimal()
```

```r
# Step 7: PCA Scatter Plot
pca_df <- data.frame(pca_result$x, Species = iris$Species)

ggplot(pca_df, aes(x = PC1, y = PC2, color = Species)) +
  geom_point(size = 3, alpha = 0.7) +
  labs(title = "PCA of Iris Dataset", x = "PC1", y = "PC2") +
  theme_minimal()
```

```r
# Step 8: Contribution of Variables to Principal Components
loadings <- as.data.frame(pca_result$rotation)  # Extract PCA loadings
loadings$Variable <- rownames(loadings)

# PC1 Contributions
ggplot(loadings, aes(x = reorder(Variable, abs(PC1)), y = abs(PC1))) +
  geom_bar(stat = "identity", fill = "steelblue", color = "black") +
  coord_flip() +
  labs(title = "Variable Contribution to PC1", x = "Variable", y = "Absolute Contribution") +
  theme_minimal()

# PC2 Contributions
ggplot(loadings, aes(x = reorder(Variable, abs(PC2)), y = abs(PC2))) +
  geom_bar(stat = "identity", fill = "steelblue", color = "black") +
  coord_flip() +
  labs(title = "Variable Contribution to PC2", x = "Variable", y = "Absolute Contribution") +
  theme_minimal()
```

```r
# Step 9: Biplot (PCA Loadings and Data Points)
ggplot() +
  geom_point(data = pca_df, aes(x = PC1, y = PC2, color = Species), size = 3, alpha = 0.7) +
  geom_segment(data = loadings, aes(x = 0, y = 0, xend = PC1 * 2, yend = PC2 * 2),
               arrow = arrow(length = unit(0.2, "cm")), color = "red") +
  geom_text(data = loadings, aes(x = PC1 * 2, y = PC2 * 2, label = Variable),
            vjust = -0.5, hjust = 0.5, color = "red") +
  labs(title = "PCA Biplot of Iris Dataset", x = "PC1", y = "PC2") +
  theme_minimal()
```

```r
# Install and load required packages
install.packages("ggplot2")
install.packages("corrplot")
library(ggplot2)
library(corrplot)
```

```r

# Step 1: Load and Explore the Data
data("mtcars")  # Load the dataset
head(mtcars)  # View first few rows
summary(mtcars)  # Summary statistics
```

```r
# Step 2: Check for Missing Values
sum(is.na(mtcars))  # Should be 0 if data is clean
```

```r
# Step 3: Standardization (Mean = 0, Variance = 1)
mtcars_scaled <- scale(mtcars)  # Standardize the numerical data
```

```r
# Step 4: Perform PCA
pca_mtcars <- prcomp(mtcars_scaled, center = TRUE, scale. = TRUE)
```

```r
# Step 5: View PCA Results
summary(pca_mtcars)  # Variance explained by each component
pca_mtcars$rotation  # Loadings (weights of original variables)
pca_mtcars$sdev^2  # Eigenvalues (variance explained by each PC)
```

```r

# Step 6: Scree Plot (Variance Explained) using ggplot2
variance_explained <- pca_mtcars$sdev^2 / sum(pca_mtcars$sdev^2)
scree_df <- data.frame(PC = paste0("PC", 1:length(variance_explained)), Variance = variance_explained * 100)

ggplot(scree_df, aes(x = PC, y = Variance)) +
  geom_bar(stat = "identity", fill = "steelblue", color = "black") +
  geom_text(aes(label = round(Variance, 1)), vjust = -0.5, size = 5) +
  labs(title = "Scree Plot (Variance Explained)", x = "Principal Components", y = "Percentage of Variance") +
  theme_minimal()
```

```r
# Step 7: Correlation Matrix of Original Features
corrplot(cor(mtcars), method = "circle", type = "lower", tl.col = "black", title = "Feature Correlation", mar = c(0, 0, 1, 0))
```

```r
# Step 8: PCA Biplot using ggplot2
pca_df <- as.data.frame(pca_mtcars$x)  # Extract PCA-transformed data
pca_loadings <- as.data.frame(pca_mtcars$rotation)  # Extract loadings

# Add car names to PCA Data
pca_df$cars <- rownames(mtcars)

ggplot(pca_df, aes(x = PC1, y = PC2, label = cars)) +
  geom_point(color = "blue", size = 3) +
  geom_text(aes(label = cars), vjust = -0.5, size = 3) +
  labs(title = "PCA Biplot of mtcars Dataset", x = "PC1", y = "PC2") +
  theme_minimal()

# Overlay PCA Loadings
ggplot() +
  geom_segment(data = pca_loadings, aes(x = 0, y = 0, xend = PC1 * 2, yend = PC2 * 2), arrow = arrow(length = unit(0.2, "cm")), color = "red") +
  geom_text(data = pca_loadings, aes(x = PC1 * 2, y = PC2 * 2, label = rownames(pca_loadings)), vjust = -0.5, hjust = 0.5, color = "red") +
  labs(title = "PCA Loadings", x = "PC1", y = "PC2") +
  theme_minimal()
```

```r
# Step 9: Contribution of Variables to PC1
loadings_df <- data.frame(Variables = rownames(pca_mtcars$rotation), Contribution = abs(pca_mtcars$rotation[,1]))

ggplot(loadings_df, aes(x = reorder(Variables, -Contribution), y = Contribution)) +
  geom_bar(stat = "identity", fill = "lightblue", color = "black") +
  coord_flip() +
  labs(title = "Contribution of Variables to PC1", x = "Variables", y = "Absolute Contribution") +
  theme_minimal()
```

```r
# Step 10: Scatter Plot of PCA-transformed Data
ggplot(pca_df, aes(x = PC1, y = PC2, label = cars)) +
  geom_point(color = "blue", size = 3) +
  geom_text(vjust = -0.5, size = 3) +
  labs(title = "PCA Scatter Plot of mtcars Dataset", x = "PC1", y = "PC2") +
  theme_minimal()
```
